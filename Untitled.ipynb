```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import IsolationForest
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
from imblearn.under_sampling import RandomUnderSampler
```

```python
data = pd.read_csv('data/creditcard.csv')
data.head()
```

```python
data.info()
```

No Missing Data as seen above

```python
data['Class'].value_counts() / len(data) * 100
```

Only 0.17% of the data is fraud. Hence, highly imbalanced dataset

```python
ss = StandardScaler()
data['Amount'] = ss.fit_transform(data[['Amount']])
data['Time'] = ss.fit_transform(data[['Time']])
```

```python
data_X = data.drop(['Class', 'Time'], axis = 1)
data_Y = data['Class']
X_train, X_test, y_train, y_test = train_test_split(data_X, data_Y, stratify = data_Y, test_size = 0.2)
```

```python
y_test.value_counts()
```

```python
rus = RandomUnderSampler()
X_under, y_under = rus.fit_resample(X_train, y_train)
```

```python
f, ax2 = plt.subplots(1, 1, figsize=(20,15))
sub_sample_corr = X_under.corr()
sns.heatmap(sub_sample_corr, cmap='coolwarm_r', annot_kws={'size':20}, ax=ax2)
ax2.set_title('SubSample Correlation Matrix \n (use for reference)', fontsize=14)
plt.show()
```
